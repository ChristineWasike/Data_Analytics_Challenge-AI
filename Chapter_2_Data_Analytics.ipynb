#### Step 6: Create a pivot table and feed it with data

Now that our data is loaded to the dataframe and we have described its statistical information, let us convert it to JSON with the "record" orientation specified. The resulting list-like structure is the data format that the pivot table understands [3].

```python
json_data = data.to_json(orient='records')
```

Define a pivot table object with the report configuration specified [3]:

```python
pivot_table_configuration = {
    "container": "#pivot-container",
    "width": "100%",
    "height": 600,
    "toolbar": True,
    "report": {
        "dataSource": {
            "type": "json",
            "data": json.loads(json_data)
        },
        "slice": {
            "reportFilters": [{
                    "uniqueName": "Calendar Year"
                },
                {
                    "uniqueName": "Month"
                }
            ],
            "rows": [
                {
                "uniqueName": "Month"
            }
            ],
            "columns": [
                {
                    "uniqueName": "Calendar Year"
                }],
            "measures": [{
                    "uniqueName": "Tax Revenue"
                },
            ],
            "sorting": {
                "column": {
                    "type": "desc",
                    "tuple": [],
                    "measure": "Tax Revenue"
                }
            }
        },
        "options": {
            "showAggregationLabels": False
        },
        "formats": [{
            "name": "",
            "thousandsSeparator": " ",
            "decimalSeparator": ".",
            "decimalPlaces": 2,
            "currencySymbol": "$",
            "currencySymbolAlign": "left",
            "nullValue": "",
            "textAlign": "right",
            "isPercent": False
        }]
    }
}     
```

So far, we have configured the following:

* A data slice 

This is a subset of the fields we want to present on the grid

* A data source and its type

Using json.loads() method, we deserialized a string containing a JSON document to a Python object.

* Optional view-related settings

That is, conditional formatting, and numerical values formatting.

Now let us convert the pivot table presented by a Python dictionary to a JSON formatting string [3]:

```python
webdatarocks_json_object = json.dumps(pivot_table_configuration)
```

#### Step 7: Render the pivot table in HTML

This is the most important step as here, we will define a function that will render the pivot table in the notebook cell.

```python
def render_table_and_charts(webdatarocks_json_object, draw_charts=''''''):
    
    code = '''
      <link href="https://cdn.webdatarocks.com/latest/webdatarocks.min.css" rel="stylesheet"/>
      <script src="https://cdn.webdatarocks.com/latest/webdatarocks.js"></script>
      <script src="https://www.gstatic.com/charts/loader.js"></script>
      <h1 style="margin-bottom: 20px;">Monthly United States - Tax Revenue</h1>
      <div id="pivot-container"></div>
      <script>
      var pivot = new WebDataRocks({0});
      
      {1}
      </script>
      '''.format(webdatarocks_json_object, draw_charts)
    return HTML(code)
```

So far, we have specified the HTML layout that contains the scripts of WebDataRocks, CSS styles, and containers where the pivot table will be rendered.

The entire layout is enclosed in triple quotes.

Additionally, we pass an argument that stands for executing custom JavaScript code. This is where we implement the logic of how and when the chart should be drawn. This should happen only when the pivot table is fully rendered, loaded with data, and localization files. We can track this with the "reportcomplete" event.

Here is the code that gets the data from the table, creates a chart, and passes the summarized data to it [3]:

```python
# JS code for drawing charts on the pivot report loading completion

draw_charts = ''' 
var pivotTableReportComplete = false;

pivot.on("reportcomplete", () => {
    pivot.off("reportcomplete");
    pivotTableReportComplete = true;
});
    
'''
```

Finally, let us call the rendering function and see the resulting dashboard:

As you will see, the pivot table is filled with data and the fields are displayed according to the report rules we set. 

Notice that the dashboard itself is interactive. You can play around with different elements of the reporting tool, slice and dice the data records on the grid, filter, and sort them [3]. 

The options for filtering are 'Month' and 'Calendar Year'. This gives you a quick understanding of different tax revenue collection trends across different months and years, with the option to select the columns you want represented in the pivot table. 

Additionally, the chart reacts to every change made in the report [3].

In the pivot table below, we are looking at tax revenue collected in the months of May, June, and July, in the years 2018 and 2019 for the Petroleum and Special Fuel products by the US government. The month of June seems to have pretty high tax revenues, while Petroleum has higher revenues than Special Fuel.

```python
render_table_and_charts(webdatarocks_json_object, draw_charts)
```

Now that you have a reporting tool that allows you to look at data from multiple angles, you can keep exploring the data by changing tracked metrics, filtering the data, or try adding other charting visualizations.

##### Plot 2: Scatterplot

A great way to visualize the relationship between two continuous variables is a scatter plot. Below, we plot a scatter plot that shows the "Room" tax revenues according to the "Calendar Year". We can see there is some correlation as the points have a similar trend in that "July" had far much higher tax revenues compared to the months of "January" and "February" [7].

```python
import seaborn as sns
sns.set_style("whitegrid")

plt.figure(figsize=(15, 10))
sns.scatterplot(x="Calendar Year", y="Room",hue = 'Month',data=data,)
plt.title('Calendar Year by Room Occupancy Distribution')
```

##### Plot 3: Histogram

From the scatter plot above, we have seen how the months of January and February have really low tax revenues. So we can be sure that tax revenues will drop during these months and we need to maximize on tax collection. But what about the 'middle' months. 

To further assess the 'middle' month trends, let's plot a histogram that shows the tax revenues of a particular column, in this case, we will use the same column, 'Room', and we will also use the October, November, and December months [4].

```python
x1 = list(data[data.Month == 'October']['Room'])
x2 = list(data[data.Month == 'November']['Room'])
x3 = list(data[data.Month == 'December']['Room'])

sns.set()
plt.figure(figsize=(10,6))
plt.hist([x1,x2,x3],bins = 5,stacked=True,color = ['#337ab7','#f16e0c','#b73333'],)
plt.xlabel('Room')
plt.ylabel('Tax Revenue')
plt.legend(['October', 'November', 'December'])
plt.title('Room Occupancy Tax Revenue distribution')
```

In the first bar you notice November had very little to no tax revenues, but the second bar has very little to no tax revenues in October and December. These bars that fall between 0 and 0.8 are very low.

However, the next two bars are competing with December showing more tax revenues, followed by November, and finally December. The last bar falling between 1.2 and 1.4 has little to no returns in the month of December.

Based on the above plot, we can conclude that Room Occupancy has rapid tax revenues of the values between (0.8 and 1.2).

##### Plot 4: Bar Graph

We can also further visualize the trend in tax revenues from individual columns as shown below, that is, the 'Hospitals' tax revenues collected monthly.

This bar graph gives us an overall idea of which months has the highest and lowest, while showing those that did not have any tax revenues collected. As seen from the below graph, August, September, December, and March had very minimal tax revenues [4].

Let's start with a simple example. Using the first observation description as the input for the wordcloud, the three steps are:

* Extract the review 
* Create and generate a wordcloud image
* Display the cloud using matplotlib [6].

```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# Start with one review
text = sentiment_data.text[3]

stopwords = set(STOPWORDS)
stopwords.update(["the", "a", "VirginAmerica", "airline", "United", "flight", "fly"])

# Create and generate a word cloud image
wordcloud=WordCloud(stopwords=stopwords, max_font_size=50, max_words=100, background_color="white").generate(text)

# Display the generated image
fig = plt.figure(1)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
fig.savefig("firstreview.png", dpi=900)
```

You can see that this specific review has the expressions aggressive, obnoxious, and blast, standing out the most.

Now, let's combine all reviews into one big text and create a big fat cloud to see which characteristics are most common in these sentiment reviews by customers [6].

```python
text = " ".join(review for review in sentiment_data.text)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list
stopwords = set(STOPWORDS)
stopwords.update(["the", "a", "airline", "United", "flight", "fly"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white", max_font_size=20, max_words=20000).generate(text)

# Display the generated image using matplotlib
fig = plt.figure(figsize=(15,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
fig.savefig("allreviews.png", dpi=900)
```

We can see the overall competitiveness between USAirway and AmericanAir as well as SouthwestAir. Time also seems to stand out greatly as words like "delayed", "minute", and "waiting" are visible from the word cloud. "Thank" also stands out that speaks highly of the customers' appreciation for the flight. This is the value is qualitative data. You can affirm your hypotheses through actual customer reviews from visualizing their sentiments in a word cloud [6].
